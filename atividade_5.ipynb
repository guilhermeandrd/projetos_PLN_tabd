```python
def preprocess(text: str) -> list:
    # text: texto bruto de entrada
    # retorno: lista de tokens de text
    text = text.lower()
    text = text.replace(',','')
    text = text.replace('.','')
    text = text.replace('\n','')

    tokens = text.split(' ')
    tokens = [token for token in tokens if token]
    return tokens

def build_vocab(tokens: list) -> dict:
    # tokens: lista de palavras do corpus (todos os documentos)
    # retorno: {'words': list, 'word2idx': dict}, onde words é indexa
    # todas as palavras do corpus e word2idx indexa um dicionário que
    # mapeia cada palavra a um valor numérico inteiro

    unique_words = sorted(list(set(tokens)))
    vocab = {'words': unique_words, 'word2idx': {}}

    for i, word in enumerate(unique_words):
      vocab['word2idx'][word] = i

    return vocab

def compute_tf(tokens: list, vocab: dict) -> list:
    # tokens: lista de palavras de um documento
    # vocab: {'words': list, 'word2idx': dict} construido por build_vocab
    # retorno: vetor de frequências (lista)
    tf_vector = [0] * len(vocab['words'])
    for token in tokens:
        if token in vocab['word2idx']:
            tf_vector[vocab['word2idx'][token]] += 1

    return tf_vector

def compute_tfidf(docs_tokens: list, vocab: dict) -> list:
    # docs_tokens: lista de documentos, cada um sendo lista de tokens
    # vocab: {'words': list, 'word2idx': dict} construido por build_vocab
    # retorno: matriz tfidf (lista de listas)

    tfs = []
    for doc_tokens in docs_tokens:
        tfs.append(compute_tf(doc_tokens, vocab))

    num_docs = len(docs_tokens)
    idf_vector = [0] * len(vocab['words'])
    for i, word in enumerate(vocab['words']):
        doc_count = 0
        for doc_tokens in docs_tokens:
            if word in doc_tokens:
                doc_count += 1
        if doc_count > 0:
            idf_vector[i] = np.log(num_docs / doc_count)

    tfidf_matrix = []
    for tf_vec in tfs:
        tfidf_vec = [tf * idf for tf, idf in zip(tf_vec, idf_vector)]
        tfidf_matrix.append(tfidf_vec)

    return tfidf_matrix

def cosine_similarity(vec1: list, vec2: list) -> float:
    # vec1: vetor numérico 1
    # vec2: vetor numérico 2
    # retorno: valor numérico da similaridade

    n_vec1 = np.array(vec1)
    n_vec2 = np.array(vec2)

    norm_vec1 = np.linalg.norm(n_vec1)
    norm_vec2 = np.linalg.norm(n_vec2)

    # otimizacao para os calculos
    if norm_vec1 == 0 or norm_vec2 == 0:
        return 0.0

    cos_sim = np.dot(n_vec1, n_vec2) / (norm_vec1 * norm_vec2)
    return cos_sim
```

Usando o TF-IDF, obtenha a representação vetorial de cada documento. Calcule a similaridade dos documentos dois a dois. Quais são os documentos mais similares?

```python
def build_cooccurrence(docs_tokens: list, vocab: dict, window_size: int = 2) -> list:
    # docs_tokens: lista de documentos, cada um sendo lista de tokens
    # vocab: {'words': list, 'word2idx': dict} construido por build_vocab
    # window_size: tamanho da janela de contexto
    # retorno: matriz de coocorrência (lista de listas)

    vocab_size = len(vocab['words'])
    co_matrix = np.zeros((vocab_size, vocab_size), dtype=int)

    for doc_tokens in docs_tokens:
        if not doc_tokens:
            continue
        for i, target_word in enumerate(doc_tokens):
            target_word_idx = vocab['word2idx'].get(target_word)
            if target_word_idx is None:
                continue

            start = max(0, i - window_size)
            end = min(len(doc_tokens), i + window_size + 1)

            for j in range(start, end):
                if i == j:
                    continue
                context_word = doc_tokens[j]
                context_word_idx = vocab['word2idx'].get(context_word)
                if context_word_idx is None:
                    continue

                co_matrix[target_word_idx, context_word_idx] += 1
    return co_matrix.tolist()

def compute_ppmi(docs_tokens: list, vocab: dict, window_size: int = 2) -> list:
    # tokens: lista de palavras do corpus
    # vocab: {'words': list, 'word2idx': dict} construido por build_vocab
    # window_size: tamanho da janela de contexto
    # retorno: matriz PPMI (lista de listas)

    co_matrix_np = np.array(build_cooccurrence(docs_tokens, vocab, window_size))
    vocab_size = len(vocab['words'])

    total_cooccurrences = np.sum(co_matrix_np)

    if total_cooccurrences == 0:
        return np.zeros((vocab_size, vocab_size)).tolist()

    p_ij = co_matrix_np / total_cooccurrences

    p_i = np.sum(co_matrix_np, axis=1) / total_cooccurrences
    p_i = p_i.reshape(-1, 1)

    p_j = np.sum(co_matrix_np, axis=0) / total_cooccurrences
    p_j = p_j.reshape(1, -1)

    # evitar logaritmos de zero (que resultam em -inf), substituindo valores muito pequenos por um epsilon.
    p_ij[p_ij == 0] = 1e-10
    p_i[p_i == 0] = 1e-10
    p_j[p_j == 0] = 1e-10

    pmi_matrix = np.log2(p_ij / (p_i * p_j))

    ppmi_matrix = np.maximum(0, pmi_matrix)

    return ppmi_matrix.tolist()

def get_word_vector(word: str, vocab: dict, matrix: list) -> list:
    # word: palavra a ser consultada
    # vocab: lista de palavras únicas
    # matrix: matriz de vetores (ex: PPMI)
    # retorno: vetor que representa a palavra (lista) ou None

    word_idx = vocab['word2idx'].get(word)


    if word_idx is None:
        return None

    return matrix[word_idx]
```

```python

window_size = 2

ppmi_matrix = compute_ppmi(cleaned_docs_tokens, global_vocab, window_size=window_size)
print(f"Matriz PPMI calculada com shape: ({len(ppmi_matrix)}, {len(ppmi_matrix[0]) if ppmi_matrix else 0})")


ppmi_matrix_np = np.array(ppmi_matrix)

similarities = []
vocab_words = global_vocab['words']
vocab_size = len(vocab_words)

for i in range(vocab_size):
    for j in range(i + 1, vocab_size): # 'j' começa de 'i+1' para considerar apenas pares únicos (w1, w2) e não (w2, w1) ou (w1, w1)
        word1 = vocab_words[i]
        word2 = vocab_words[j]

        vec1 = get_word_vector(word1, global_vocab, ppmi_matrix)
        vec2 = get_word_vector(word2, global_vocab, ppmi_matrix)


        if vec1 is not None and vec2 is not None and np.sum(vec1) > 0 and np.sum(vec2) > 0:
            sim = cosine_similarity(vec1, vec2)
            if not np.isnan(sim):
                similarities.append((sim, word1, word2))

similarities.sort(key=lambda x: x[0], reverse=True)

print("\nOs 10 pares de palavras mais similares (usando PPMI e Similaridade de Cosseno):")
for sim, w1, w2 in similarities[:10]:
    print(f"  '{w1}' e '{w2}': Similaridade = {sim:.4f}")
```
